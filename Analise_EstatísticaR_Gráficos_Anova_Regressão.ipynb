```
#Verificando a normalidade dos dados com Shapiro Wilk
#H0 = os dados seguem uma distribuição normal
#H1 = os dados não seguem uma distribuição normal

#Resultado: p-valor > 0,05 (nível de significância para 95% de confiança), não rejeitando a hipótese nula, ou seja, os dados seguem uma distribuição normal

shapiro.test(dados$Preco)
```

```
#plotando o gráfico Q-Q plot para visualizar a distribuição
qqnorm(dados$Preco)
qqline(dados$Preco)
```

```
#Explorando a relação entre as variáveis Preço e Quadrimestre

#Boxplot entre o Preco e Quadrimestre
boxplot(dados$Preco~ dados$Quadrimestre)
```

```
library (ggplot2)
```

```
ggplot(data = dados, aes(x = Preco , fill = Quadrimestre)) +
geom_density(alpha = 0.4) +
xlim(200,600)
```

```
# ANOVA (analysis of variance) utiliza um teste F para identificar se ha variação significativa ao realizar as comparações de média de n populações. 
# Permite comparar mais de 2 média

#H0 = as médias são iguais
#H1 = pelo menos uma das médias é diferente

#Realizando a analise de variancia por ANOVA
anova <- aov(Preco ~ Quadrimestre, data = dados)
summary(anova)
```

```
#Explorando a relação entre as variáveis Preço e Portas

boxplot(dados$Preco ~ dados$Portas)

#Test t de Student
t.test(dados$Preco ~ dados$Portas , 
       paired = FALSE, #amostras nao pareadas
       alternative = 'two.sided', #bilateral
       conf.level = 0.95 #95% de confianca
       )
```

```
ggplot(data = dados, aes(x = Preco , fill = Portas)) +
geom_density(alpha = 0.4) +
xlim(200,600)
```

```
# função de densidade com variáveis aleatórias rt de 20 amostras
n = 20
plot(density(rt(n = n, df = n-1)))
```

```
#Explorando a relação entre as variáveis Preço e Quilometragem

plot(y = dados$Preco ,
     x = dados$Quilometragem,
     pch = 16)
```
